```python
# function to define trend based on Day_perc_change where n is the parameter
def trend(n):
    if(n < 0.5 and n >-0.5):
        return 'Slight or no change'
    elif(n < 1 and n >= .5):
        return 'Slight positive'
    elif(n <= -.5 and n > -1):
        return 'Slight negative'
    elif(n < 3 and n >= 1):
        return 'Positive'
    elif(n <= -1 and n > -3):
        return 'Negative'
    elif(n < 7 and n >= 3):
        return 'Among top gainers'
    elif(n <= -3 and n > -7):
        return 'Among top losers'
    elif(n >= 7):
        return 'Bull run'
    elif(n <= -7):
        return 'Bear drop'
```

```python
df['Trend'] = df['Day_Perc_Change'].apply(trend)
```
