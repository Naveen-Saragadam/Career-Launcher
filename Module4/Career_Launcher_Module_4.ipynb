```python
# Impoting essential libraries 
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing data
df = pd.read_csv('./stock_data/RBLBANK2.csv')
df.head()
```

```python
# setting Date to index
df.set_index('Date',inplace=True,)
df.head()
```

```python
# Checking Null values
df.isna().sum() 
```

```python
# Dropping Null values
df.dropna(axis=0,inplace=True)
df.head()
```

```python
# Changing the Column names
old_names = ['Close Price']
new_names = ['Price']
df.rename(columns=dict(zip(old_names,new_names)),inplace=True)
df.head()
```

```python
# Funcion 
'''def call(df):
    if df['Price'] < df['Lower_band']:
        return 'Call'
    elif ((df['Price']>df['Lower_band'])&(df['Price']<df['Middle_band'])):
        return "Hold Buy/Liquidate Short"
    elif ((df['Price']<df['Upper_band'])&(df['Price']>df['Middle_band'])):
        return "Hold Short/Liquidate Buy"
    else:
        return 'Short'
    df.Call.value_counts()
'''
```

```python
# Creating the Call column
df['Call'] = 'Short'
df.loc[df['Price']<df['Lower_band'], 'Call'] = 'Buy'
df.loc[(df['Price']>df['Lower_band'])&(df['Price']<df['Middle_band']), 'Call'] = 'Hold Buy'
df.loc[(df['Price']>df['Middle_band'])&(df['Price']<df['Upper_band']), 'Call'] = 'Hold Short'
df['Call'].value_counts()
```

```python
df.columns
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from  sklearn import metrics 
scr = StandardScaler()
lbc = LabelEncoder()

X = df[['Middle_band','Upper_band', 'Lower_band', 'Price']]
y = df['Call']
y = lbc.fit_transform(y).reshape(-1,1)
X = scr.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

regressor.fit(X_train,y_train)
regressor.coef_
```

```python
#Testing accuracy 
y_pred = regressor.predict(X_test)
metrics.r2_score(y_test,y_pred)
```

```python
#Training accuracy
regressor.score(X_train,y_train)
```

```python
# Plotting (y_test,y_pred
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
```

```python
# Importing Classsification Algortihms
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
# Creating an OrderdDict()
max_features = y.max()
from collections import OrderedDict
classifier_models = OrderedDict([
    ("Nearest Neighbors", KNeighborsClassifier(max_features)),
    ("Linear SVM",        SVC(kernel="linear", C=0.025)),
    ("RBF SVM",           SVC(gamma=2, C=1)),
    ("Decision Tree",     DecisionTreeClassifier(max_depth=5)),
    ("Random Forest",     RandomForestClassifier(max_depth=5, n_estimators=10, max_features=max_features)),
    ("AdaBoost",          AdaBoostClassifier()),
    ("Naive Bayes",       GaussianNB())
])

classifier_models
```

```python
# Acurracy from OrderdDict()
accuracy_scores = {}
for model_name, classifier in classifier_models.items():
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    accuracy_scores[model_name] = classifier.score(X_test, y_test)
    
accuracy_scores = OrderedDict(sorted(accuracy_scores.items(), key=lambda x: x[1]))
accuracy_scores
```

```python
# Logestic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
LR = LogisticRegression()
LR.fit(X,y)
LR_pred = LR.predict(X_test)
print('F1 score:', f1_score(y_test, LR_pred,average='weighted'))
print('Acuuracy:', accuracy_score(y_test, LR_pred))
```

```python
Maruti_df = pd.read_csv('./stock_data/Maruti.csv')
Maruti_df.head()
```

```python
#computing bollinger bands 
Maruti_df['30 Day STD'] = Maruti_df['Close Price'].rolling(window=20).std()
Maruti_df['Middle_band'] = Maruti_df['Close Price'].rolling(window=20).mean()
Maruti_df['Upper_band'] = Maruti_df['Middle_band'] + (Maruti_df['30 Day STD'] * 2)
Maruti_df['Lower_band'] = Maruti_df['Middle_band'] - (Maruti_df['30 Day STD'] * 2)
```

```python
Maruti_df.dropna(inplace=True)
```

```python
#Percentage change 
Maruti_df['Price_Open_Close'] = (Maruti_df['Close Price']-Maruti_df['Open Price'])/Maruti_df['Open Price'] *100
Maruti_df['Price_Low_high'] = (Maruti_df['High Price']-Maruti_df['Low Price'])/Maruti_df['Low Price'] *100
```

```python
Maruti_df.head()
```

```python
Maruti_df['Day_Perc_Change'] = Maruti_df['Close Price'].pct_change().fillna(0)

Maruti_df['5day_mean'] = Maruti_df['Day_Perc_Change'].rolling(5).mean()
Maruti_df['5day_std'] = Maruti_df['Day_Perc_Change'].rolling(5).std()
```

```python
# creating Action column
Maruti_df['Action'] = np.nan

for i in range(len(Maruti_df['Close Price'])-1):
    if Maruti_df['Close Price'].iloc[i+1] > Maruti_df['Close Price'].iloc[i]:
        Maruti_df['Action'].iloc[i] = 1
    else :
        Maruti_df['Action'].iloc[i] = -1
```

```python
Maruti_df.dropna(inplace = True)
Maruti_df.head()
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split

X=Maruti_df[['Price_Open_Close','Price_Low_high','5day_mean','5day_std']].values
y=Maruti_df['Action'].values
X = scr.fit_transform(X)

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3)
```

```python
rfc = RandomForestClassifier(n_estimators=10, max_features=2)
rfc
```

```python

rfc.fit(X_train,  y_train)
print(rfc.score(X_test, y_test))
print(rfc.score(X_train,  y_train))
```

```python
# Cumulative Product of PCT change in Close_Price with predicted actions
plt.figure(figsize=(25, 7))
cumulative_returns = ( 1 +( Maruti_df['Close Price'].pct_change() * Maruti_df['Action'])).dropna().cumprod()
cumulative_returns.plot()
plt.grid(True)
```
