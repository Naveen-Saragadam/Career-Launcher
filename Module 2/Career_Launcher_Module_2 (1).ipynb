```python
# computing voltality for Day_Perc_Change 
rolling_std = df['Day_Perc_Change'].rolling(window=7).std()
rolling_std.isnull().sum()
rolling_std.dropna(inplace = True,axis=0)
```

```python
#ploting voltality of RBLBANK
plt.figure(figsize=(16,8))
plt.title('voltality of RBLBANK')
plt.plot(rolling_std)
```

```python
#Reading nifty50 index dataset
nifty = pd.read_csv("./stock_data/Nifty50.csv")
nifty.head()
```

```python
# computig per_change() for nifty50
nifpc = nifty['Close'].pct_change() * 100
```

```python
df.columns
```

```python
#Computing nifty50 voltality
nifty_voltality = nifpc.rolling(window=7).std()
nifty_voltality.index = df.index
```

```python
#comparing voltality indexs of both RBLBANK and Nifty50
plt.figure(figsize=(16,8))
plt.plot(rolling_std)
plt.plot(nifty_voltality)
plt.legend(loc='best',labels=['RBLBANK','Nifty50'])
plt.title('Voltality comparision for RBLBANK and Nitfy 50')
```

```python
price = df['Close Price']
```

```python
#defining moving average for 21 days
ma_21 = price.rolling(window=21).mean()
```

```python
#defining moving average for 34 days
ma_34 = price.rolling(window=34).mean()
```

```python
#ploting simple moving avrages 
plt.figure(figsize=(16,8))
plt.plot(price, label='Price')
plt.plot(ma_21,label='sma_21days')
plt.plot(ma_34,label='sma_34days')
plt.legend(loc='best')
plt.ylabel('Price in $')
```

```python
#computing bollinger bands 
df['30 Day STD'] = df['Close Price'].rolling(window=20).std()
df['Middle_band'] = df['Close Price'].rolling(window=20).mean()
df['Upper_band'] = df['Middle_band'] + (df['30 Day STD'] * 2)
df['Lower_band'] = df['Middle_band'] - (df['30 Day STD'] * 2)
#df['Middle_band'] = (df['Upper Band']+ df['Lower Band']) /2
```

```python
#ploting bolinger bands for RBLBANK stockdata
plt.figure(figsize=(16,8))
plt.plot(df['Close Price'],label = 'Close price',linewidth = 2.5,)
plt.plot(df['Middle_band'],label = 'Middle_band', dashes = [2, 2, 5, 2])
plt.plot(df['Upper_band'],label = 'Upper_band',dashes = [6, 2])
plt.plot(df['Lower_band'],label = 'Lower_band', dashes = [6, 2])
plt.legend(loc='best')
plt.title('Bollinger bands for RBLBANK stock data')
```
